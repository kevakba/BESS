#### Data Loading

```python
# target variable
price = pd.read_csv('/home/kevin/Downloads/BESS/data/data_June2024_Dec2024/price_01062024_31122024.csv')

# independent variables
ail = pd.read_csv('/home/kevin/Downloads/BESS/data/data_June2024_Dec2024/AIL_01062024_31122024.csv')
solar = pd.read_csv('/home/kevin/Downloads/BESS/data/data_June2024_Dec2024/solar_generation_june_dec_2024_2.csv')
wind = pd.read_csv('/home/kevin/Downloads/BESS/data/data_June2024_Dec2024/wind_generation_june_dec_2024_2.csv')
temp_calgary = pd.read_csv('/home/kevin/Downloads/BESS/data/data_June2024_Dec2024/temperature_calgary_01062024_31122024.csv')
temp_edmonton = pd.read_csv('/home/kevin/Downloads/BESS/data/data_June2024_Dec2024/temperature_edmonton_01062024_31122024.csv')
temp_fortmc = pd.read_csv('/home/kevin/Downloads/BESS/data/data_June2024_Dec2024/temperature_fortmc_01062024_31122024.csv')
```

```python
display(price.head())

display(ail.head())
display(solar.head())
display(wind.head())
display(temp_calgary.head())
display(temp_edmonton.head())
display(temp_fortmc.head())
```

```python
price.drop(['Unnamed: 0'], axis=1, inplace=True)
ail.drop(['Unnamed: 0'], axis=1, inplace=True)
solar.drop(['Unnamed: 0'], axis=1, inplace=True)
wind.drop(['Unnamed: 0'], axis=1, inplace=True)
temp_calgary.drop(['Unnamed: 0'], axis=1, inplace=True)
temp_edmonton.drop(['Unnamed: 0'], axis=1, inplace=True)
temp_fortmc.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
display(price.head())

display(ail.head())
display(solar.head())
display(wind.head())
display(temp_calgary.head())
display(temp_edmonton.head())
display(temp_fortmc.head())
```

```python
price.rename(columns={'begin_datetime_mpt': 'datetime_', 'rolling_30day_avg': 'rolling_30day_avg_price'}, inplace=True)

ail.rename(columns={'begin_datetime_mpt': 'datetime_'}, inplace=True)

solar.rename(columns={'Date (MPT)': 'datetime_', 'Volume':'solar_generation'}, inplace=True)

wind.rename(columns={'Date (MPT)': 'datetime_', 'Volume':'wind_generation'}, inplace=True)

temp_calgary.rename(columns={'Timestamp_mst': 'datetime_', 'Temperature (degree C)':'temp_calgary'}, inplace=True)

temp_edmonton.rename(columns={'Timestamp_mst': 'datetime_', 'Temperature (degree C)':'temp_edmonton'}, inplace=True)

temp_fortmc.rename(columns={'Timestamp_mst': 'datetime_', 'Temperature (degree C)':'temp_fortmc'}, inplace=True)
```

```python
display(price.head())

display(ail.head())
display(solar.head())
display(wind.head())
display(temp_calgary.head())
display(temp_edmonton.head())
display(temp_fortmc.head())
```

```python
# convert columns to datetime
price['datetime_'] = pd.to_datetime(price['datetime_'])
ail['datetime_'] = pd.to_datetime(ail['datetime_'])
solar['datetime_'] = pd.to_datetime(solar['datetime_'])
wind['datetime_'] = pd.to_datetime(wind['datetime_'])
temp_calgary['datetime_'] = pd.to_datetime(temp_calgary['datetime_'])
temp_edmonton['datetime_'] = pd.to_datetime(temp_edmonton['datetime_'])
temp_fortmc['datetime_'] = pd.to_datetime(temp_fortmc['datetime_'])
```

```python
# Perform outer merge on all dataframes on 'datetime_'
merged_df = price.merge(ail, on='datetime_', how='outer') \
                 .merge(solar, on='datetime_', how='outer') \
                 .merge(wind, on='datetime_', how='outer') \
                 .merge(temp_calgary, on='datetime_', how='outer') \
                 .merge(temp_edmonton, on='datetime_', how='outer') \
                 .merge(temp_fortmc, on='datetime_', how='outer')

# Display the merged dataframe
display(merged_df.head())
```
