```python
# convert columns to datetime
price['datetime_'] = pd.to_datetime(price['datetime_'])
ail['datetime_'] = pd.to_datetime(ail['datetime_'])
solar['datetime_'] = pd.to_datetime(solar['datetime_'])
wind['datetime_'] = pd.to_datetime(wind['datetime_'])
temp_calgary['datetime_'] = pd.to_datetime(temp_calgary['datetime_'])
temp_edmonton['datetime_'] = pd.to_datetime(temp_edmonton['datetime_'])
temp_fortmc['datetime_'] = pd.to_datetime(temp_fortmc['datetime_'])
ws_calgary['datetime_'] = pd.to_datetime(ws_calgary['datetime_'])   
ws_edmonton['datetime_'] = pd.to_datetime(ws_edmonton['datetime_'])
ws_fortmc['datetime_'] = pd.to_datetime(ws_fortmc['datetime_'])
```

```python
# Perform outer merge on all dataframes on 'datetime_'
merged_df = price.merge(ail, on='datetime_', how='outer') \
                 .merge(solar, on='datetime_', how='outer') \
                 .merge(wind, on='datetime_', how='outer') \
                 .merge(temp_calgary, on='datetime_', how='outer') \
                 .merge(temp_edmonton, on='datetime_', how='outer') \
                 .merge(temp_fortmc, on='datetime_', how='outer') \
                 .merge(ws_calgary, on='datetime_', how='outer') \
                 .merge(ws_edmonton, on='datetime_', how='outer') \
                 .merge(ws_fortmc, on='datetime_', how='outer')

# Display the merged dataframe
display(merged_df.head())
```

#### time based variables

```python
merged_df['hour_of_day'] = merged_df['datetime_'].dt.hour
merged_df['day_of_week'] = merged_df['datetime_'].dt.dayofweek
merged_df['week_of_month'] = merged_df['datetime_'].dt.isocalendar().week
merged_df['month'] = merged_df['datetime_'].dt.month

# Display the updated dataframe
display(merged_df.head())
```

#### seasonality

```python
# This must be unordered categorical variable since we dont know which season cause price to rise

# Define a function to determine the season
def get_season(month):
    if month in [5, 6, 7, 8, 9]:
        return 'summer'
    else:
        return 'winter'

# Apply the function to create a new column 'season'
merged_df['season'] = merged_df['month'].apply(get_season)

# Display the updated dataframe
display(merged_df.head())
```

#### pool_price_lagged_values_till_24_hours

```python
def generate_lagged_features(df, column, num_lags):
    for lag in range(1, num_lags + 1):
        df[f'{column}_lag_{lag}'] = df[column].shift(lag)
    return df

# Generate 24-hour lagged values for the pool_price column
merged_df = generate_lagged_features(merged_df, 'pool_price', 24)

# Display the updated dataframe
display(merged_df.head())
```

#### AIL_lagged_values_till_24_hours

```python
# Generate 24-hour lagged values for the AIL column
merged_df = generate_lagged_features(merged_df, 'alberta_internal_load', 24)

# Display the updated dataframe
display(merged_df.head())
```
